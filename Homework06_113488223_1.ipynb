```python
# Section 2 - Chained Conditionals
# Start the script by creating and initializing a variable with a value of your choice
var_entry = input("Please enter a value of your choice: ") # We take input as a string
# Based on the data type (type()) of the variable, write a chained conditional that prints the data type to the screen:
var_check = var_entry.isalpha() # 1st Check: If the input is composed entirely of letters
if var_check is True:
    print ("This is a string!")
elif var_check is False: # 2nd Check: If the input is not composed of letters, then we evaluate the data type further
    var_type = eval(var_entry)
    if type(var_type) is int:
        print ("This is an integer!")
    elif type(var_type) is float:
        print ("This is a float!")
```

```python
# Section 3 - User Input and Conditionals
# Write a script that uses the age of the fair visitor as input and prints the admission as output to the screen
age_input = input("Please enter the age of the Oklahoma State Fair attendee: ")
age_check = age_input.isnumeric() # check that the input is only composed of numbers

if age_check is True:
    age_number = int(age_input) # ages are in years, integer data is useful for the calculation of gate prices
    if age_number <= 5:
        print ("The price of admission is: Free!")
    elif age_number >= 6 and age_number <= 11:
        print ("The price of admission is: $6")
    if age_number >= 12:
        print ("The price of admission is: $12")
elif age_check is False: # Only numbers are allowed in age entries
    print ("Please enter the attendee's age in a proper format and try again")
```
